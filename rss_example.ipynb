```python
# RSS ÏôÄ ÎπÑÏä∑
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def get_dom_path(tag):
    """ÌÉúÍ∑∏Ïùò CSS ÏÑ†ÌÉùÏûê Í≤ΩÎ°ú ÏÉùÏÑ±"""
    path = []
    while tag is not None and tag.name != '[document]':
        name = tag.name
        # idÍ∞Ä ÏûàÎã§Î©¥ idÎ°ú ÏãùÎ≥Ñ
        if tag.has_attr('id'):
            name += f"#{tag['id']}"
        # classÍ∞Ä ÏûàÎã§Î©¥ classÎ°ú Ï∂îÍ∞Ä
        elif tag.has_attr('class'):
            class_str = "." + ".".join(tag['class'])
            name += class_str
        path.insert(0, name)
        tag = tag.parent
    return " > ".join(path)

# BBC Î©îÏù∏ ÌéòÏù¥ÏßÄ ÏöîÏ≤≠
base_url = "https://www.bbc.com/"
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Ï§ëÎ≥µ Ï†úÍ±∞Ïö©
seen = set()

print("üìå ÏàòÏßëÎêú ÎßÅÌÅ¨ Î™©Î°ù:\n")

for a in soup.find_all('a', href=True):
    title = a.get_text(strip=True)
    href = a['href']

    if not title or href.startswith('#'):
        continue

    full_url = urljoin(base_url, href)
    dom_path = get_dom_path(a)

    if 'main-content' not in dom_path:
        continue

    key = (title, full_url)
    if key not in seen:
        seen.add(key)
        print(f"Title: {title}")
        print(f"URL: {full_url}")
        # print(f"üìé DOM Path: {dom_path}\n")
        print(a)
```
